```python
import pandas as pd
from sklearn.datasets import load_boston

boston = load_boston()
# boston # 
df = pd.DataFrame(boston["data"], columns = boston["feature_names"]) # pandas
df["ans"] = boston["target"]
df
print("-----------------------")
import numpy as np
from sklearn.model_selection import train_test_split

x = df.drop("ans", axis = 1) #  "ans",  x,  df  ans 
x = np.array(x)
y = np.array(df["ans"])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1) # 90% x, 10% x, 90% y, 10% y
show = x_test
print(show)
print("-----------------------")
x_train

d = pd.DataFrame([
        [1, 2],
        [3, 4]
    ], columns = ["a", "b"], index = ["a", "b"])

d.drop("a", axis = 0) # b(3, 4)  a , 
d.drop("a", axis = 1) # b(2, 4)  a , 
print("-----------------------")
from sklearn.tree import DecisionTreeRegressor #  Classifier 

reg = DecisionTreeRegressor(max_depth = 2) # 
reg.fit(x_train, y_train)

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize = (20, 20)) # , , 
plot_tree(reg, 
          feature_names = boston["feature_names"], 
          filled = True) 
```
